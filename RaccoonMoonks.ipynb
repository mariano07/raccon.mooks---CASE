2. Faça uma requisição ao arquivo JSON desenvolvido para retornar os dados de
funcionários.

```python
import requests

url = "https://storage.googleapis.com/raccoon-humane/psel.json"

response = requests.get(url)

if response.status_code == 200:
    funcionarios = response.json()
    print(funcionarios)
else:
    print("Não foi possível obter os dados dos funcionários.")
```

cargo e salário de um funcionário

```python
for funcionario in funcionarios:
    nome = funcionario.get("nome")
    cpf = funcionario.get("cpf")
    cargo = funcionario.get("cargo")
    salario = funcionario.get("salario")

    print("Nome: ", nome)
    print("CPF: ", cpf)
    print("Cargo: ", cargo)
    print("Salário: ", salario)
    print("\n")
```
